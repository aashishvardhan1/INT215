## Standard Calorie Intake:

Age b/w 0-2 years: 800 calories, 
Age b/w 2-4 years: 1400 calories, 
Age b/w 4-8 years: 1800 calories.

```python
# Standard calorie Intake as per age
if age >= 4 and age < 8:
    std_cal = 1800
elif age >= 2:
    std_cal = 1400
elif age > 0:
    std_cal = 800
    
print(f"Your required intake of calories: {std_cal}")
```

## Calories per 100g of Food

Milk: 100 cal/ 100g, 
Egg: 155 cal/ 100g, 
Rice: 130 cal/ 100g, 
Lentils: 113 cal/ 100g, 
Vegetable: 85 cal/100g, 
Meat: 143 cal/ 100g.

```python
print("Now enter your daily food consumption details \n (Enter 0 if you are not consuming)")
```

```python
# Taking input of the intake of food by the child

milk = int(input("Enter your consumption of Milk per day in grams: "))
egg = int(input("Enter your consumption of Egg per day in grams: "))
rice = int(input("Enter your consumption of Rice per day in grams: "))
lentils = int(input("Enter your consumption of Lentils per day in grams: "))
veggies = int(input("Enter your consumption of Vegetables per day in grams: "))
meat = int(input("Enter your consumption of Meat per day in grams: "))
```

```python
# Calculate the actual intake of calories by the child
intake_calories = milk + egg*1.55 + rice*1.3 + lentils*1.13 + veggies*0.85 + meat*1.43
```

```python
if intake_calories >= std_cal:
    print("Yay!, You are properly nourished")
else:
    print('Beware! You are undernourished! Take Care of your diet')
```

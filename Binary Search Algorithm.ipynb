```python
# Creating a list of random numbers between 0 to 100, such that every succeeding number having a difference of 2 between them.
# Here we are taking upper boundary and lower boundary as input from user itself (As mentioned in the problem statement)

# The user can also directly replace the existing list with a own new list.

lower = int(input("Enter the Lower boundary of your array: "))
upper = int(input("Enter the Upper boundary of your array: "))
arr = list(range(lower,upper,2))
```

```python
# Taking a random number from user as input
x = int(input("Enter a random number to search: "))
```

```python
# Function that performs binary search
def binarySearchAlgo(list1, first, last, x):
    
    if last >= first:
        mid = first + (last - first) // 2

        if list1[mid] == x:
            return mid
        
        elif list1[mid] > x:
            return binarySearchAlgo(list1, first, mid-1, x)

        else:
            return binarySearchAlgo(list1, mid + 1, last, x)
 
    else:
        return -1
```

```python
# Storing the result of the search
result = binarySearchAlgo(arr, 0, len(arr)-1, x)
```

```python
if result == -1:
    print('Element Not Found in the provided list')
else:
    print(f"The element '{x}' was found at index {result} of the list")
```
